```python
class Solution:
    def kidsWithCandies(self, candies: list[int], extraCandies: int) -> list[bool]:
        results = []
        maxcandies=max(candies)
        for i in range(len(candies)) :
            results.append(maxcandies<=candies[i] + extraCandies)
        return results
        
```

```python
run = Solution()
```

```python
run.kidsWithCandies([12,1,12],10)
```
