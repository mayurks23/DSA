```python
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        sum = 0
        prod = 1
        while n > 0:
            r = n%10
            n = n//10
            sum += r
            prod *= r
        return prod - sum
```

```python
run = Solution()
```

```python
run.subtractProductAndSum(234)
```
